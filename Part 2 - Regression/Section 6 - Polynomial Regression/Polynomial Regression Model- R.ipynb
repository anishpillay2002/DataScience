```r
# Importing the dataset
dataset = read.csv('Position_Salaries.csv')

dataset = dataset[2:3];
```

```r
head(dataset,6)
```

```r
# Splitting the dataset into the Training set and Test set
library(caTools)
set.seed(123)
```

```r
# Linear Regression
lin_reg = lm(formula = Salary ~ ., data = dataset)
```

```r
# Fitting Polynomial Regression to the dataset
dataset$Level2= dataset$Level ^2;
dataset$Level3= dataset$Level ^3;
dataset$Level4= dataset$Level ^4;
head(dataset,6)
```

```r
# Polynomial Regression
poly_reg = lm(formula = Salary ~ ., data = dataset)
```

```r
summary(poly_reg)
```

```r
install.packages('ggplot2')
```

```r
library(ggplot2)
```

```r
# Visualising the Linear regression
ggplot() +
geom_point(aes(x = dataset$Level, y =dataset$Salary), color ='red') +
geom_line(aes(x = dataset$Level, y = predict(lin_reg,newdata = dataset)), color ='green') +
          ggtitle('Truth or Bluff : Linear Regression') +
          xlab('Level') +
          ylab('Salary')
```

```r
# Visualising the Polynomial regression
ggplot() +
geom_point(aes(x = dataset$Level, y =dataset$Salary), color ='red') +
geom_line(aes(x = dataset$Level, y = predict(poly_reg,newdata = dataset)), color ='green') +
          ggtitle('Truth or Bluff : Polynomial Regression') +
          xlab('Level') +
          ylab('Salary')
```

```r
# Predicting the salary according to Linear Regression
y_pred = predict(lin_reg, data.frame(Level = 6.5))
print(y_pred)
```

```r
# Predicting the salary according to Linear Regression
y_poly_pred = predict(poly_reg, data.frame(Level = 6.5,Level2 = 6.5^2,Level3 = 6.5^3,Level4 = 6.5^4))
print(y_poly_pred)
```
