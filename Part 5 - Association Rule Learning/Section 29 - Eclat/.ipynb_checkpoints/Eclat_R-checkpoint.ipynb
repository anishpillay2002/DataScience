## Eclat vs Apriori Usage
Use Apriori, when you need to know dependence of one product on the other. eg: When a customer buys milk , he also buys cereal
Use Eclat, when you need to know the occurence of group of items. eg:  Customer buys mineral water, milk together always

```r
# Eclat

# Data Preprocessing
# install.packages('arules')
library(arules)
dataset = read.csv('Market_Basket_Optimisation.csv')
dataset = read.transactions('Market_Basket_Optimisation.csv', sep = ',', rm.duplicates = TRUE)
summary(dataset)
```

```r
itemFrequencyPlot(dataset, topN = 10)
```

```r

# Training Eclat on the dataset
rules = eclat(data = dataset, parameter = list(support = 0.003, minlen = 2))
```

```r

# Visualising the results
inspect(sort(rules, by = 'support')[1:10])
```
